````python
import numpy as np
````

````python
x = np.array([[1,2]]).T
b = np.array([[1,1]]).T
W1=np.array([[3,1],[2,4]])
````

````python
z = np.dot(W1,x)  +  b
z
````

````python
W2 = np.array([[3,2],[5,1]])
````

````python
y = np.dot(W2,z) + b
y
````
