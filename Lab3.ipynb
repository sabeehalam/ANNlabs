## Binary Classification problem where a perceptron with linear activation function is not effective.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

points = 200
```

```python
x1 = np.random.randn(points)-2
x2 = np.random.randn(points)-2
plt.scatter(x1,x2,c=("coral"))
x3 = np.random.randn(points)+2
x4 = np.random.randn(points)+2
plt.scatter(x3,x4,c=("lightblue"))
plt.title('Binary Classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

```python
X1 = np.concatenate((x1,x3))
X2 = np.concatenate((x2,x4))
```

```python
w = np.random.randn(2,1)
plt.plot(X1*w[0], X1*w[1] + 1, linestyle='solid')
plt.scatter(x1,x2,c=("coral"))
plt.scatter(x3,x4,c=("lightblue"))
plt.title('Binary Classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

```python
x1 = np.random.randn(points)-2
x2 = np.random.randn(points)-2
plt.scatter(x1,x2,c=("coral"))

x3 = np.random.randn(points)+2
x4 = np.random.randn(points)+2
plt.scatter(x3,x4,c=("coral"))

x5 = np.random.randn(points)-2
x6 = np.random.randn(points)+2
plt.scatter(x5,x6,c=("lightblue"))

x7 = np.random.randn(points)+2
x8 = np.random.randn(points)-2
plt.scatter(x7,x8,c=("lightblue"))

plt.title('Binary Classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

```python
X1 = np.concatenate((x1,x3,x5,x7))
X2 = np.concatenate((x2,x4,x6,x8))
```

```python
w = np.random.randn(2,1)
plt.plot(X1*w[0], X1*w[1] + 1, linestyle='solid')
plt.scatter(x1,x2,c=("coral"))
plt.scatter(x3,x4,c=("coral"))
plt.scatter(x5,x6,c=("lightblue"))
plt.scatter(x7,x8,c=("lightblue"))
plt.title('Binary Classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```
